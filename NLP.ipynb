# Removing stop words
 Removing stop words,or common English words that do not change the sentiment of a sentence, is the first step. By removing them, the sentiment analysis should run faster, but not be less accurate (as the stop words do not affect sentiment, but they do slow down the analysis).

 **Stop words are available in abundance in any human language. By removing these words, we remove the low-level information from our text in order to give more focus to the important information.Feb 10, 2021**


```python
# remove stop words - can be slow for a lot of text!
nltk.download('stopwords')
start = time.time()
cache = set(stopwords.words("english"))
def remove_stopwords(review):
  text = " ".join([word for word in review.split() if word not in cache])
  return text

# remove the stop words from both columns
df2.Negative_Review = df2.Negative_Review.apply(remove_stopwords)
df2.Positive_Review = df2.Positive_Review.apply(remove_stopwords)
```

```python
df2.head()
```

```python
# Create the vader sentiment analyser ( there are others in NLTK you can try too)

vader_sentiment = SentimentIntensityAnalyzer()

# There are 3 possibilities of input for a review:
# It could be "No Negative", in which case, return 0 
# It could be "No Positive", in which case, return 0
# It could be a review, in which case calculate the sentiment
def calc_sentiment(review):
  if review == "No Positive" or review == "No Negative":
    return 0
  return vader_sentiment.polarity_scores(review)["compound"]
```

```python
# Add a negative sentiment and positive sentiment column

print("Calculating sentiment columns for both positive and negative reviews"
)

start = time.time()
df2["Negative_Sentiment"] = df2.Negative_Review.apply(calc_sentiment)
df2["Positive_Sentiment"] = df2.Positive_Review.apply(calc_sentiment)
end = time.time()
print("Calculating sentiment took " + str(round(end -start,2)) + " Seconds" )
```

```python
df2 = df2.sort_values(by=["Negative_Sentiment"], ascending=True)
print(df2[["Negative_Review", "Negative_Sentiment"]])
df2 = df2.sort_values(by=["Positive_Sentiment"], ascending=True)
print(df2[["Positive_Review", "Positive_Sentiment"]])
```
